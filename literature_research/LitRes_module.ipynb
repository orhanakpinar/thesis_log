```python
import pandas as pd



class Annotation:
    
    columns = ["Title", "Citation", "In_Text_Citation", "..."]

    Agro_econ = "Agro_econ"
    Agro_international = "Agro_international"
    Agro_policy = "Agro_policy"
    Agro_tech = "Agro_tech"
    Agroecology =  "Agroecology"
    Autonomy = "Autonomy"
    Big_agro = "Big_agro"
    Buyuksehir_Law = "Buyuksehir_Law"
    Collectives = "Collectives"
    Cooperatives = "Cooperatives"
    Debt = "Debt"
    Deruralization = "Deruralization"
    Education = "Education"
    Food_Network = "Food_Network"
    Food_Security =  "Food_Security"
    Food_Sovereignty = "Food_Sovereignty"
    Gender =  "Gender"
    Health = "Health"
    History = "History"
    Interdisciplinary = "Interdisciplinary"
    Land_Consolidation = "Land_Consolidation"
    Land_Policy = "Land_Policy"
    Land_Use = "Land_Use"
    Methodology = "Methodology"
    Migration =  "Migration"
    Monoculture = "Monoculture"
    Monoculture_Poli = "Monoculture_Poli"
    Policy_Access = "Policy_Access"
    Risks_Global = "Risks_Global"
    Rural_Development = "Rural_Development"
    Rural_Family =  "Rural_Family"
    Rural_Livelihood = "Rural_Livelihood" 
    Rural_Policy =  "Rural_Policy"
    Rurban = "Rurban"
    Seed =  "Seed"
    Shortfood = "Shortfood"
    Small_holder = "Small_holder"
    Survivorship_bias = "Survivorship_bias"
    TR_agroEcon = "TR_agroEcon" 
    TR_agroGov = "TR_agroGov"
    TR_landUse = "TR_landUse"
    TR_ruralGov = "TR_ruralGov"
    TR_Peasant = "TR_Peasant"
    Urbanization = "Urbanization"
    Variable = "Variable"
    
    def __init__(self):
        self.Title = ""
        self.Citation = ""
        self.In_Text_Citation = ""
        self.Agro_econ = ""
        self.Agro_international = ""
        self.Agro_policy = ""
        self.Agro_tech = ""
        self.Agroecology = ""
        self.Autonomy = ""
        self.Big_agro = ""
        self.Buyuksehir_Law = ""
        self.Collectives = ""
        self.Cooperatives = ""
        self.Debt = ""
        self.Deruralization = ""
        self.Education = ""
        self.Food_Network = ""
        self.Food_Security = ""
        self.Food_Sovereignty = ""
        self.Gender = ""
        self.Health = ""
        self.History = ""
        self.Interdisciplinary = ""
        self.Land_Consolidation = ""
        self.Land_Policy = ""
        self.Land_Use = ""
        self.Methodology = ""
        self.Migration = ""
        self.Monoculture = ""
        self.Monoculture_Poli = ""
        self.Policy_Access = ""
        self.Risks_Global = ""
        self.Rural_Development = ""
        self.Rural_Family = ""
        self.Rural_Livelihood = ""
        self.Rural_Policy = ""
        self.Rurban = ""
        self.Seed = ""
        self.Shortfood = ""
        self.Small_holder = ""
        self.Survivorship_bias = ""
        self.TR_agroEcon = ""
        self.TR_agroGov = ""
        self.TR_landUse = ""
        self.TR_ruralGov = ""
        self.TR_Peasant = ""
        self.Urbanization = ""
        self.Variable = ""

    def add_annotation(self, attribute, annotation):
        if hasattr(self, attribute):
            current_value = getattr(self, attribute)
            if current_value:
                updated_value = f'{current_value},"{annotation}"'
            else:
                updated_value = f'"{annotation}"'
            setattr(self, attribute, updated_value)
        else:
            raise AttributeError(f"Attribute {attribute} does not exist in Annotation class.")



    def add_annots_from_dict(self, annotations_dict):
    """
    Add annotations to the provided Annotation object from a dictionary.
    
    Parameters:
    - annotation_obj: Instance of Annotation class
    - annotations_dict: Dictionary where keys are annotation texts and values are lists of categories
    """
        for annotation_text, categories in annotations_dict.items():
            for category in categories:
                self.add_annotation(category, annotation_text)

    def add_to_excel(self, filename):
        # Load existing Excel file
        df = pd.read_excel(filename)

        # Add a new row for the current instance
        new_row = {attr: getattr(self, attr) or 0 for attr in self.__dict__.keys()}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

        # Save back to Excel file
        df.to_excel(filename, index=False)




```
