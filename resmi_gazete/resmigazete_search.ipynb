```python
#IRRELEVANT; only for apriori research

import pandas as pd
import re
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt

def preprocess_text(text):
    text = re.sub(r'[^a-zA-ZçğıöşüÇĞİÖŞÜ\s]', '', str(text))
    text = text.lower()
    return text

def generate_embeddings(texts, model, tokenizer):
    embeddings = []
    for text in texts:
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
        embeddings.append(embedding)
    return torch.tensor(embeddings)

def validate_clusters(embeddings, labels, kmeans):
    silhouette_avg = silhouette_score(embeddings, labels)
    cluster_distances = []
    for i in range(kmeans.n_clusters):
        cluster_points = embeddings[labels == i]
        if len(cluster_points) > 1:
            intra_distances = cdist(cluster_points, [kmeans.cluster_centers_[i]])
            avg_distance = np.mean(intra_distances)
            cluster_distances.append(avg_distance)
        else:
            cluster_distances.append(0)
    inter_cluster_distances = cdist(kmeans.cluster_centers_, kmeans.cluster_centers_)
    np.fill_diagonal(inter_cluster_distances, np.nan)
    metrics = {
        'Silhouette Score': silhouette_avg,
        'Average Intra-Cluster Distance': np.mean(cluster_distances),
        'Minimum Inter-Cluster Distance': np.nanmin(inter_cluster_distances),
        'Average Inter-Cluster Distance': np.nanmean(inter_cluster_distances)
    }
    return metrics

def test_and_plot_clusters(file_path, cluster_list):
    df = pd.read_excel(file_path)
    df['Processed_Text'] = df['Text'].apply(preprocess_text)
    model_name = "dbmdz/bert-base-turkish-cased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    embeddings = generate_embeddings(df['Processed_Text'].tolist(), model, tokenizer).numpy()
    results = []
    best_df = None
    best_silhouette = -1
    for n_clusters in cluster_list:
        kmeans = KMeans(n_clusters=n_clusters, random_state=0)
        labels = kmeans.fit_predict(embeddings)
        df['Cluster'] = labels
        metrics = validate_clusters(embeddings, labels, kmeans)
        metrics['Number of Clusters'] = n_clusters
        results.append(metrics)
        print(f"Number of Clusters: {n_clusters} - Metrics: {metrics}")
        if metrics['Silhouette Score'] > best_silhouette:
            best_silhouette = metrics['Silhouette Score']
            best_df = df.copy()
    results_df = pd.DataFrame(results)
    # Plot metrics
    plt.figure(figsize=(12, 6))
    plt.plot(results_df['Number of Clusters'], results_df['Silhouette Score'], marker='o', label='Silhouette Score')
    plt.plot(results_df['Number of Clusters'], results_df['Average Intra-Cluster Distance'], marker='o', label='Average Intra-Cluster Distance')
    plt.plot(results_df['Number of Clusters'], results_df['Average Inter-Cluster Distance'], marker='o', label='Average Inter-Cluster Distance')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Metric Value')
    plt.title('Clustering Validation Metrics vs. Number of Clusters')
    plt.legend()
    plt.grid(True)
    plt.show()
    # Return the best DataFrame
    return best_df

#def save_best_cluster_to_excel(df, file_name):
#    df.to_excel(file_name, index=False)

# Usage example:
file_path = 'resmigazete_tarım_filter.xlsx'
cluster_list = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 80, 90, 100, 105, 110, 120, 130, 150, 170, 180]
test_and_plot_clusters(file_path, cluster_list)
```

```python
#IRRELEVANT; only for apriori research

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
import re
import nltk

# Download necessary NLTK resources
nltk.download('stopwords')
from nltk.corpus import stopwords

def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', str(text))
    text = text.lower()
    stop_words = set(stopwords.words('turkish'))  # Adjust language as needed
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

def cluster_texts(file_path, num_clusters=5):
    df = pd.read_excel(file_path)
    if 'Text' not in df.columns:
        raise ValueError("The file must contain a 'Text' column.")
    df['Processed_Text'] = df['Text'].apply(preprocess_text)
    tfidf_vectorizer = TfidfVectorizer(max_features=1000)
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['Processed_Text'])
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    labels = kmeans.fit_predict(tfidf_matrix)
    df['Cluster'] = labels
    terms = tfidf_vectorizer.get_feature_names_out()
    cluster_keywords = {}
    for i in range(num_clusters):
        cluster_center = kmeans.cluster_centers_[i]
        top_terms_indices = cluster_center.argsort()[-10:][::-1]
        top_terms = [terms[ind] for ind in top_terms_indices]
        cluster_keywords[i] = top_terms
    df['Cluster_Keywords'] = df['Cluster'].map(cluster_keywords)
    metrics = validate_clusters(tfidf_matrix.toarray(), labels, kmeans)
    metrics['Number of Clusters'] = num_clusters
    return df, cluster_keywords, metrics

def validate_clusters(embeddings, labels, kmeans):
    silhouette_avg = silhouette_score(embeddings, labels)
    cluster_distances = []
    for i in range(kmeans.n_clusters):
        cluster_points = embeddings[labels == i]
        if len(cluster_points) > 1:
            intra_distances = cdist(cluster_points, [kmeans.cluster_centers_[i]])
            avg_distance = np.mean(intra_distances)
            cluster_distances.append(avg_distance)
        else:
            cluster_distances.append(0)
    inter_cluster_distances = cdist(kmeans.cluster_centers_, kmeans.cluster_centers_)
    np.fill_diagonal(inter_cluster_distances, np.nan)
    metrics = {
        'Silhouette Score': silhouette_avg,
        'Average Intra-Cluster Distance': np.mean(cluster_distances),
        'Minimum Inter-Cluster Distance': np.nanmin(inter_cluster_distances),
        'Average Inter-Cluster Distance': np.nanmean(inter_cluster_distances)
    }
    return metrics

def test_and_plot_clusters(file_path, cluster_list):
    results = []
    for n_clusters in cluster_list:
        _, _, metrics = cluster_texts(file_path, n_clusters)
        results.append(metrics)
        print(f"Number of Clusters: {n_clusters} - Metrics: {metrics}")
    results_df = pd.DataFrame(results)
    plt.figure(figsize=(12, 6))
    plt.plot(results_df['Number of Clusters'], results_df['Silhouette Score'], marker='o', label='Silhouette Score')
    plt.plot(results_df['Number of Clusters'], results_df['Average Intra-Cluster Distance'], marker='o', label='Average Intra-Cluster Distance')
    plt.plot(results_df['Number of Clusters'], results_df['Average Inter-Cluster Distance'], marker='o', label='Average Inter-Cluster Distance')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Metric Value')
    plt.title('Clustering Validation Metrics vs. Number of Clusters')
    plt.legend()
    plt.grid(True)
    plt.show()

# Usage example:
file_path = 'resmigazete_tarım_filter.xlsx'
cluster_list = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 80, 90, 100, 105, 110, 120, 130, 150, 170, 180]
test_and_plot_clusters(file_path, cluster_list)
```

```python
#IRRELEVANT; only for apriori research

import pandas as pd
import numpy as np
import re
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.cluster import KMeans
from collections import Counter

def preprocess_text(text):
    text = re.sub(r'[^a-zA-ZçğıöşüÇĞİÖŞÜ\s]', '', str(text))
    text = text.lower()
    return text

def generate_embeddings(texts, model, tokenizer):
    embeddings = []
    for text in texts:
        inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
        embeddings.append(embedding)
    return np.array(embeddings)

def extract_cluster_keywords(df, num_keywords=5):
    """
    Extracts top keywords for each cluster based on simple frequency analysis.
    
    Parameters:
    - df (pd.DataFrame): Dataframe containing text and cluster assignments.
    - num_keywords (int): Number of keywords to extract per cluster.
    
    Returns:
    - dict: Dictionary mapping cluster labels to lists of top keywords.
    """
    keywords = {}
    for cluster in df['Cluster'].unique():
        cluster_text = ' '.join(df[df['Cluster'] == cluster]['Processed_Text'])
        words = cluster_text.split()
        common_words = [word for word, _ in Counter(words).most_common(num_keywords)]
        keywords[cluster] = common_words
    return keywords

def cluster_texts_with_bert(file_path, num_clusters=5):
    df = pd.read_excel(file_path)
    if 'Text' not in df.columns:
        raise ValueError("The file must contain a 'Text' column.")
    
    df['Processed_Text'] = df['Text'].apply(preprocess_text)

    model_name = "dbmdz/bert-base-turkish-cased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)

    embeddings = generate_embeddings(df['Processed_Text'].tolist(), model, tokenizer)

    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    df['Cluster'] = kmeans.fit_predict(embeddings)

    cluster_keywords = extract_cluster_keywords(df)
    df['Cluster_Keywords'] = df['Cluster'].map(cluster_keywords)

    return df, cluster_keywords

# Usage example:
file_path = 'resmigazete_tarım_filter.xlsx'
num_clusters = 50
processed_df, cluster_keywords = cluster_texts_with_bert(file_path, num_clusters)

# Sort and save the processed dataframe with cluster info and keywords
processed_df = processed_df.sort_values(by='Cluster')
processed_df.to_excel("BERT_cluster.xlsx", index=False)
```

```python
from sklearn.metrics import silhouette_score
import pandas as pd
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

# Load the relevant-only dataset with predicted clusters
df = pd.read_excel('BERT_complete_relevant_only_BERTopic_clusters.xlsx')

# Ensure the "Cluster" column exists
if 'Cluster' not in df.columns:
    raise ValueError("The file must contain a 'Cluster' column for validation.")

# Ensure the "Text" column exists
if 'Text' not in df.columns:
    raise ValueError("The file must contain a 'Text' column for validation.")

# Initialize the SentenceTransformer model for Turkish
embedding_model = SentenceTransformer("dbmdz/bert-base-turkish-cased")

# Generate embeddings for the "Text" column
print("Generating embeddings...")
embeddings = embedding_model.encode(df['Text'].tolist(), show_progress_bar=True)

# Initialize BERTopic
topic_model = BERTopic(language="turkish", calculate_probabilities=True)

# Fit the BERTopic model using precomputed embeddings
print("Training BERTopic model...")
topics, probs = topic_model.fit_transform(df['Text'].tolist(), embeddings)

# Retrieve the cluster labels
labels = topics

# Calculate the Silhouette Score
silhouette_avg = silhouette_score(embeddings, labels)
print(f"Silhouette Score: {silhouette_avg:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Load the specified Excel file into a DataFrame
file_path = 'BERT_relevant_clusters_with_agrotopic_annotation_filtered_year.xlsx'
df = pd.read_excel(file_path)
df = df[df["Year"]<=2020]

# Group data by AgroPolicy_Topic and count occurrences of each year
grouped = df.groupby(["AgroPolicy_Topic", "Year"]).size().reset_index(name='Counts')

# Create separate plots for AgroPolicy_Topic = 0 and AgroPolicy_Topic = 1
for topic in [0, 1]:
    topic_data = grouped[grouped['AgroPolicy_Topic'] == topic]
    
    # Set figure size wider and taller for better visibility
    plt.figure(figsize=(15, 8))
    
    # Create the bar plot
    plt.bar(topic_data['Year'], topic_data['Counts'], width=0.6)
    
    # Add title and labels with larger font size
    plt.title(f"Yearly Counts for AgroPolicy_Topic = {topic}", fontsize=16)
    plt.xlabel("Year", fontsize=14)
    plt.ylabel("Counts", fontsize=14)
    
    # Set ticks with larger font size and increased spacing
    plt.xticks(topic_data['Year'].unique(), fontsize=12, rotation=45, ha='right')
    plt.yticks(fontsize=12)
    
    # Add grid lines for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Increase spacing between ticks
    plt.tight_layout()
    
    # Display the plot
    plt.show()
```

```python
import pandas as pd

supports_list_generalLaw = [8, 23, 24, 25, 27, 28]
supports_list_farmerSupport = [0, 2, 3, 9, 12, 13, 15, 18, 19, 20]
supports_list_investment = [11, 17]
supports_list_financial = [1, 14]
supports_list_negative = [6, 20, 26]
supports_list_logistics = [5, 7, 16]
supports_list_damageLoss = [4, 10, 21]

agropolicy_agreements = pd.read_excel("Final_AgroPolicy_Topic_Agreements_Annotated.xlsx")
agropolicy_supports = pd.read_excel("Final_AgroPolicy_Topic_Supports_Annotated.xlsx")

agropolicy_agreements = agropolicy_agreements[(agropolicy_agreements["Year"] <= 2020)& (agropolicy_agreements["Year"] >= 2004)]
agropolicy_supports = agropolicy_supports[(agropolicy_supports["Year"] <= 2020) & (agropolicy_supports["Year"] >= 2004)]

pd.set_option('display.max_rows', None)  # Show all rows



agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_generalLaw), 
    "Annotation_Group"] = 0

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_farmerSupport), 
    "Annotation_Group"] = 1

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_investment), 
    "Annotation_Group"] = 2

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_financial), 
    "Annotation_Group"] = 3

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_negative), 
    "Annotation_Group"] = 4

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_logistics), 
    "Annotation_Group"] = 5

agropolicy_supports.loc[
    agropolicy_supports["Annotation_Topic"].isin(supports_list_damageLoss), 
    "Annotation_Group"] = 6

# Group and print results
print(agropolicy_agreements["Annotation_Topic"].value_counts())
print(agropolicy_supports["Annotation_Group"].value_counts())
print(agropolicy_supports["Annotation_Topic"].value_counts())
```

```python
import matplotlib.pyplot as plt
bins = [2004, 2008, 2012, 2016, 2020]
labels = ["2004–2007", "2008–2011", "2012–2015", "2016–2019"]

# Example bins covering 2005–2020 by 4-year intervals:

# Create a new column 'Year_Group' in your DataFrame
agropolicy_agreements["Year_Group"] = pd.cut(
    agropolicy_agreements["Year"], 
    bins=bins, 
    labels=labels, 
    right=False
)
agropolicy_supports["Year_Group"] = pd.cut(
    agropolicy_supports["Year"], 
    bins=bins, 
    labels=labels, 
    right=False
)


# Now group by 'Year_Group' instead of just 'Year'
agreement_counts = agropolicy_agreements.groupby("Year_Group")["Annotation_Topic"].value_counts()
support_counts = agropolicy_supports.groupby("Year_Group")["Annotation_Group"].value_counts()

# Plot each Annotation_Topic for agreements, using Year_Group instead of Year
for topic in agreement_counts.index.get_level_values("Annotation_Topic").unique():
    topic_data = agreement_counts.xs(topic, level="Annotation_Topic")
    plt.figure(figsize=(10, 6))
    plt.bar(topic_data.index, topic_data.values, color='skyblue', label=f"Agreements - {topic}")
    plt.title(f"Agropolicy Agreements: {topic}")
    plt.xlabel("Year Group")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot each Annotation_Group for supports, again using Year_Group
for topic in support_counts.index.get_level_values("Annotation_Group").unique():
    topic_data = support_counts.xs(topic, level="Annotation_Group")
    plt.figure(figsize=(10, 6))
    plt.bar(topic_data.index, topic_data.values, color='orange', label=f"Supports - {topic}")
    plt.title(f"Agropolicy Supports: {topic}")
    plt.xlabel("Year Group")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
