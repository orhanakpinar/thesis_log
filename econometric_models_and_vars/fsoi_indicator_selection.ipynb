```python
# Character Encoding
replacements = {
    'Ý': 'İ',
    'ý': 'ı',
    'ð': 'ğ',
    'þ': 'ş',
    'Þ': 'Ş'
}

data_TÜİK["Location_Name"] = data_TÜİK["Location_Name"].str.translate(str.maketrans(replacements))
data_TÜİK["Location_Name"] = data_TÜİK["Location_Name"].str.capitalize()
data_TÜİK["Location_Name"].unique()
```

```python
data_TÜİK["Mean_Household_Size"] = data_TÜİK["Mean_Household_Size"].str.replace(',', '.').astype(float)
data_TÜİK = data_TÜİK.replace("-", "0")
data_TÜİK = data_TÜİK.convert_dtypes()
# Force all numeric columns to float64
numeric_cols = data_TÜİK.select_dtypes(include=['number']).columns
data_TÜİK[numeric_cols] = data_TÜİK[numeric_cols].astype('float64')
data_TÜİK["Year"] = data_TÜİK["Year"].astype("int64")
data_TÜİK.info()
```

```python
data_TÜİK["Greenhouse_Agro_LandUse_Decare"] = data_TÜİK["Greenhouse_Agro_LandUse_Decare"].astype(float)
data_TÜİK["Greenhouse_Agro_Production_Ton"] = data_TÜİK["Greenhouse_Agro_Production_Ton"].astype(float)
data_TÜİK["Agriculture_LandUse_Fallow_Hectare"] = data_TÜİK["Agriculture_LandUse_Fallow_Hectare"].astype(float)
data_TÜİK["Electric_Energy_Use_MWh"] = data_TÜİK["Electric_Energy_Use_MWh"].astype(float)
data_TÜİK["Agriculture_LandUse_Vegetables_Hectare"] = data_TÜİK["Agriculture_LandUse_Vegetables_Hectare"].astype(float)
data_TÜİK["Water_Refined_1000m3PerYear"] = data_TÜİK["Water_Refined_1000m3PerYear"].astype(float)
data_TÜİK.info()
```

```python
# Data from https://www.harita.gov.tr/il-ve-ilce-yuzolcumleri

data_HGM = pd.read_excel("Türkiye_Municipal_Areas.xlsx")
data_HGM.head(10)
```

```python
# Deleting town area values, keeping only city area values

data_HGM = data_HGM[
    data_HGM["İL VE İLÇE YÜZ ÖLÇÜMLERİ"].isna() &
    data_HGM["Unnamed: 1"].isna() &
    data_HGM["Unnamed: 2"].notna()]

data_HGM.reset_index(inplace=True, drop=True)
print(data_HGM)
```

```python
import unicodedata

# Correction of city name values for Türkiye-HGM data

data_HGM = data_HGM.rename(columns={"Unnamed: 2" : "Location_Name", 
                        "Unnamed: 3" : "Area_km2"})

data_HGM = data_HGM.drop(columns=["İL VE İLÇE YÜZ ÖLÇÜMLERİ", "Unnamed: 1"])
data_HGM.drop(index=0, axis=0, inplace=True)
data_HGM.reset_index(inplace=True, drop=True)


# Capitalize city names while preserving Turkish encoding
def normalize_turkish(text):
    if not isinstance(text, str):
        return text
    # Normalize Unicode composition (fixes dotted İ etc.)
    text = unicodedata.normalize("NFKC", text.strip())
    # Replace locale-specific characters before capitalization
    text = text.translate(str.maketrans("IİŞĞÜÇÖ", "ıişğüçö")).capitalize()
    return text

# Apply to city names
data_HGM["Location_Name"] = data_HGM["Location_Name"].apply(normalize_turkish)


print(data_HGM["Location_Name"].unique())
# For Hakkari, İstanbul, and İzmir
data_HGM["Location_Name"] = data_HGM["Location_Name"].str.replace("Hakkâri", "Hakkari")
data_HGM["Location_Name"] = data_HGM["Location_Name"].str.replace("Istanbul", "İstanbul")
data_HGM["Location_Name"] = data_HGM["Location_Name"].str.replace("Izmir", "İzmir")
print(data_HGM["Location_Name"].unique())

```

```python
data_official_Türkiye = data_TÜİK.merge(
    data_HGM.assign(Location_lower=data_HGM["Location_Name"].str.lower())[["Location_lower", "Area_km2"]],
    left_on=data_TÜİK["Location_Name"].str.lower(),
    right_on="Location_lower",
    how="left"
).drop(columns="Location_lower")
data_official_Türkiye["Area_km2"] = data_official_Türkiye["Area_km2"].astype("float64")
```

```python
# Control empty area values

missing_area = data_official_Türkiye[data_official_Türkiye["Area_km2"].isna()]
print(missing_area["Location_Name"].unique())

# Because that I replaced the city encodings, only Türkiye was left, picked a reference year 2008 

total_area = data_official_Türkiye.loc[
    (data_official_Türkiye["Year"] == 2008) & (data_official_Türkiye["Location_Name"] != "Türkiye"), "Area_km2"].sum()

# Assign total to Türkiye
data_official_Türkiye.loc[data_official_Türkiye["Location_Name"] == "Türkiye", "Area_km2"] = total_area

missing_area = data_official_Türkiye[data_official_Türkiye["Area_km2"].isna()]
print(missing_area["Location_Name"].unique())


```

```python
law6360_cities = ["Aydın", "Balıkesir", "Denizli", "Hatay", "Malatya", "Manisa", "Mardin", "Muğla", "Kahramanmaraş", "Ordu", "Tekirdağ", "Trabzon", "Şanlıurfa", "Van"]
data_official_Türkiye["Treated"] = data_official_Türkiye["Location_Name"].isin(law6360_cities).astype(int)
data_official_Türkiye.columns
```

```python
# Transformations

# Household number and city are will be denominators for food sovereignty analysis.
# Exchange rates for USD/TRY taken from National Bank of the Republic of Türkiye with mean of sale and offer. 
# Exchange rates https://www.tcmb.gov.tr/kurlar/kurlar_tr.html

data_exchangerates = pd.DataFrame({
    'Year': [2008, 2010, 2012, 2014, 2016, 2018],
    'Exchange_Parity_USDTRY': [1.5245, 1.5413, 1.7819, 2.329, 3.5223, 5.2857]})

data_official_Türkiye = data_official_Türkiye.merge(
    data_exchangerates, on="Year", how="left")


data_official_Türkiye["Mean_Household_Count"] = data_official_Türkiye["Population_Total"] / data_official_Türkiye["Mean_Household_Size"]
data_official_Türkiye["Landuse_Harvest_km2"] = data_official_Türkiye["Landuse_Harvest_Hectare"] * 0.01
data_official_Türkiye["Greenhouse_Agro_LandUse_km2"] = data_official_Türkiye["Greenhouse_Agro_LandUse_Decare"] * 0.001
data_official_Türkiye["LandUse_Sowed_km2"] = data_official_Türkiye["LandUse_Sowed_Hectare"] * 0.01
data_official_Türkiye["Agriculture_LandUse_Fallow_km2"] = data_official_Türkiye["Agriculture_LandUse_Fallow_Hectare"] * 0.01
data_official_Türkiye["Agriculture_LandUse_Vegetables_km2"] = data_official_Türkiye["Agriculture_LandUse_Vegetables_Hectare"] * 0.01
data_official_Türkiye["Agriculture_Longterm_Crops_km2"] = data_official_Türkiye["Agriculture_Longterm_Crops_Hectare"] * 0.01

data_official_Türkiye["Agricultural_Production_1000USD"] = (data_official_Türkiye["Agricultural_Production_1000TL"] / data_official_Türkiye["Exchange_Parity_USDTRY"])


data_official_Türkiye = data_official_Türkiye.drop(columns=["Population_Total", "Mean_Household_Size", "Landuse_Harvest_Hectare", "Greenhouse_Agro_LandUse_Decare", 
                                                            "LandUse_Sowed_Hectare", "Agriculture_LandUse_Fallow_Hectare", "Agriculture_LandUse_Vegetables_Hectare", 
                                                            "Agriculture_Longterm_Crops_Hectare", "Agricultural_Production_1000TL", "Exchange_Parity_USDTRY"])

```

```python
# Transformations on Food Sovereignty, for each location, Per Household and Per Area metrics
# Create data for processing


# data_official_Türkiye["Mean_Household_Count"]
# data_official_Türkiye["Area_km2"]



# Market
data_official_Türkiye["agro_prod_1000USD_perArea"] = data_official_Türkiye["Agricultural_Production_1000USD"] / data_official_Türkiye["Area_km2"]

data_official_Türkiye["agro_prod_1000USD_perHousehold"] = data_official_Türkiye["Agricultural_Production_1000USD"] / data_official_Türkiye["Mean_Household_Count"]


# Production
data_official_Türkiye["agro_greenhouse_prod_ton_perArea"] = data_official_Türkiye["Greenhouse_Agro_Production_Ton"] / data_official_Türkiye["Area_km2"]

data_official_Türkiye["agro_greenhouse_prod_ton_perHousehold"] = data_official_Türkiye["Greenhouse_Agro_Production_Ton"] / data_official_Türkiye["Mean_Household_Count"]

# Water 
data_official_Türkiye["water_drainage_perArea"] = data_official_Türkiye["Water_Drainage_1000m3PerYear"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["water_refined_perArea"] = data_official_Türkiye["Water_Refined_1000m3PerYear"] / data_official_Türkiye["Area_km2"]

data_official_Türkiye["water_drainage_perHousehold"] = data_official_Türkiye["Water_Drainage_1000m3PerYear"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["water_refined_perHHousehold"] = data_official_Türkiye["Water_Refined_1000m3PerYear"] / data_official_Türkiye["Mean_Household_Count"]

# Waste
data_official_Türkiye["wasteCollected_1000ton_perArea"] = data_official_Türkiye["Waste_Collected_1000Ton"] / data_official_Türkiye["Area_km2"]

data_official_Türkiye["wasteCollected_1000ton_perHousehold"] = data_official_Türkiye["Waste_Collected_1000Ton"] / data_official_Türkiye["Mean_Household_Count"]

# Energy
data_official_Türkiye["electricty_mwh_perArea"] = data_official_Türkiye["Electric_Energy_Use_MWh"] / data_official_Türkiye["Area_km2"]

data_official_Türkiye["electricty_mwh_perHousehold"] = data_official_Türkiye["Electric_Energy_Use_MWh"] / data_official_Türkiye["Mean_Household_Count"]



# Land Use
data_official_Türkiye["landuse_harvested_perArea"] = data_official_Türkiye["Landuse_Harvest_km2"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["landuse_longtermCrops_perArea"] = data_official_Türkiye["Agriculture_Longterm_Crops_km2"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["landuse_greenhouse_perArea"] = data_official_Türkiye["Greenhouse_Agro_LandUse_km2"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["landuse_sowed_perArea"] = data_official_Türkiye["LandUse_Sowed_km2"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["landuse_fallow_perArea"] = data_official_Türkiye["Agriculture_LandUse_Fallow_km2"] / data_official_Türkiye["Area_km2"]
data_official_Türkiye["landuse_vegetables_perArea"] = data_official_Türkiye["Agriculture_LandUse_Vegetables_km2"] / data_official_Türkiye["Area_km2"]


data_official_Türkiye["landuse_harvested_perHousehold"] = data_official_Türkiye["Landuse_Harvest_km2"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["landuse_longtermCrops_perHousehold"] = data_official_Türkiye["Agriculture_Longterm_Crops_km2"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["landuse_greenhouse_perHousehold"] = data_official_Türkiye["Greenhouse_Agro_LandUse_km2"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["landuse_sowed_perHousehold"] = data_official_Türkiye["LandUse_Sowed_km2"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["landuse_fallow_perHousehold"] = data_official_Türkiye["Agriculture_LandUse_Fallow_km2"] / data_official_Türkiye["Mean_Household_Count"]
data_official_Türkiye["landuse_vegetables_perHousehold"] = data_official_Türkiye["Agriculture_LandUse_Vegetables_km2"] / data_official_Türkiye["Mean_Household_Count"]


data_official_Türkiye = data_official_Türkiye.drop(columns=["Agriculture_LandUse_Vegetables_km2", "Agriculture_LandUse_Fallow_km2", "LandUse_Sowed_km2", "Greenhouse_Agro_LandUse_km2",
                                                            "Agriculture_Longterm_Crops_km2", "Landuse_Harvest_km2", "Electric_Energy_Use_MWh", "Waste_Collected_1000Ton", 
                                                            "Water_Refined_1000m3PerYear", "Water_Drainage_1000m3PerYear", "Greenhouse_Agro_Production_Ton", 
                                                            "Agricultural_Production_1000USD", "Mean_Household_Count", "Area_km2"])


data_official_Türkiye.columns
```
